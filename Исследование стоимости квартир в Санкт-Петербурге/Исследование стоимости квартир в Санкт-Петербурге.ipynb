## Предобработка данных

```python
# Удаление пустых значений в столбцах "количество этажей" и "название населенного пункта",
# а в столбцах "кол-во балконов" и "количество дней размещения объявлений" замена на 0,
# чтоб не возникло проблем с изменением типов данных на int
df = df.dropna(subset = ['floors_total', 'locality_name'])
df['balcony'] = df['balcony'].fillna(value = 0)
df['days_exposition'] = df['days_exposition'].fillna(value = 0)

col = ['last_price'
       , 'floors_total'
       , 'balcony'
       , 'days_exposition'
       , 'total_area'
       , 'ceiling_height'
       , 'living_area'
       , 'kitchen_area'
       , 'airports_nearest'
       , 'cityCenters_nearest'
       , 'parks_around3000'
       , 'parks_nearest'
       , 'ponds_around3000'
       , 'ponds_nearest'
]
for i in col:
    df[i] = pd.to_numeric(df[i], errors = 'coerce')


df[['last_price','floors_total','balcony','days_exposition']] = df[
                                                                    ['last_price'
                                                                     ,'floors_total'
                                                                     ,'balcony'
                                                                     ,'days_exposition']
                                                                        ].astype('int')

# изменение типа данных столбца "дата публикации" во временной тип данных и округение до дней,
# т.к. нас интересует дата, а не время
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format = '%Y-%m-%dT%H:%M:%S')
df['first_day_exposition'] = df['first_day_exposition'].dt.round('1D')
df['locality_name'] = df['locality_name'].astype('str')


print('-------------------------------------')
print('Проверка типов данных после изменения')
print('-------------------------------------')
df.info()
```

```python
df
```

```python
print('----------------------')
print('До обработки пропусков')
print('----------------------')
print()
pd.DataFrame(round((df.isna().mean()*100),2)).style.background_gradient('coolwarm')
```

```python
print('-------------------------')
print('После обработки пропусков')
print('-------------------------')
# заполнение пропусков медианой по группировке.
df['ceiling_height'] = (
    df['ceiling_height']
    .fillna(df.groupby('floors_total')['ceiling_height']
            .transform('median'))
)

df['living_area'] = (
    df['living_area']
    .fillna(df.groupby('total_area')['living_area']
            .transform('median'))
)

df['kitchen_area'] = (
    df['kitchen_area']
    .fillna(df.groupby('total_area')['kitchen_area']
            .transform('median'))
)

df = df.dropna(subset = ['living_area', 'kitchen_area'])
pd.DataFrame(round((df.isna().mean()*100),2)).style.background_gradient('coolwarm')
```

```python
#  'locality_name'к нижнему регистру
df['locality_name'] = df['locality_name'].str.lower()
# привел населенные пункты к единому виду
df['locality_name'] = df['locality_name'].str.replace('ё','е')
df['locality_name'] = df['locality_name'].str.replace('поселок', 'п.')
df['locality_name'] = df['locality_name'].str.replace('деревня', 'д.')
df['locality_name'] = df['locality_name'].str.replace('городского', 'г.')
df['locality_name'] = df['locality_name'].str.replace('городской', 'г.')
df['locality_name'] = df['locality_name'].str.replace('типа', 'т.')
df['locality_name'] = df['locality_name'].str.replace('село', 'с.')
df['locality_name'] = df['locality_name'].str.replace('садовое товарищество', 'с. т.')
df['locality_name'] = df['locality_name'].str.strip()
df['locality_name'].unique()
```

### Результат предобработки данных

Пропущенные значения в столбце "число балконов" и "количество дней размещения объявления" заменил на **0**, т.к. невозможно эти данные заполнить ни логически ни статистически. Пропуски по столбцу "число балконов"  составляли почти половину датафрейма, и удалять такие строки нельзя, иначе не с чем будет работать. Такое количество пропусков возможно из-за того, что эта информация вносилась пользователями, и попросту забывали вносить.<br/>
Только столбец "высота потолков" заполнил медианой по одноименному стобцу, т.к. сложно поймать зависимость от других столбцов.<br/>
Пропуски в столбцах, где данные заполняются автоматически на основе картографических данных оставил пустыми (только столбец "расстояние до центра" изменил тип данных на **float**). Многие столбцы не участвуют в анализе, заполнять медианой или нулями нет логического обоснования.
Пропуски в столбцах с площадями заменил медианой по группировке со стобцом "total_area".<br/>
Типы данных изменил согласно следующей логике:<br/> 1) данные, относящиеся к метражу квартир оставил в тип **"float"** - стоимость квартиры зависит от цены квадратного метра, поэтому, если площадь оставлять без дробной части, то получим ненастоящую цену объекта;<br/> 2) столбец **"first_day_exposition"** изменен в тип данных **"date"**;<br/> 3) булевы значения оставил без изменений;<br/>
В столбце **locality_name** **Ё** заменил на **Е**, перевел к нижнему регистру и населенные пункты привел к общему виду, справо и слево избавился от пробелов.

## Расчёты и добавление результатов в таблицу

```python
# добавление столбца стоимости кваратного метра и перевед столбца в тип данных int
df['one_sqm_price'] = (df['last_price'] / df['total_area']).astype('int')
# добавление столбцов дня недели, месяца и года публикации
df['weekday'] = df['first_day_exposition'].dt.weekday
df['month'] =  df['first_day_exposition'].dt.month
df['year'] =  df['first_day_exposition'].dt.year

# функция для категоризации этажа объекта
def floor_category(row):
    floors_total = row['floors_total']
    floor = row['floor']
    if floor == 1:
        return 'первый'
    elif floor == floors_total:
        return 'последний'
    else:
        return 'другой'
    
# добавление столбца по категоризации этажа
df['floor_category'] = df.apply(floor_category, axis = 1)
# добавление столбцов соотношения жилой площади к общей и площади кухни к жилой
# и перевед в тип данных float с 2 знаками после запятой.
df['living_area_to_total'] = (
    (df['living_area'] / df['total_area'])
    .map(lambda x: '{:,.2f}'.format(x)).astype('float')
)
df['kitchen_area_to_total'] = (
    (df['kitchen_area'] / df['total_area'])
    .map(lambda x: '{:,.2f}'.format(x)).astype('float')
)
print('-----------------------------------------------------------')
print('Готовый датафрейм для проведения исследовательского анализа')
print('-----------------------------------------------------------')
print(df.info())
df.sample(10)
```

### Результаты расчетов и добавления новых данных
Для дальнейшего анализа добавил новые столбцы:<br/>1) стоимость квадратоного метра;<br/>
2) по публикации объявления: день недели, месяц и год;<br/>
3) категорию этажа;<br/>
4) соотношение жилой площади к общей;<br/>
5) соотношение площади кухни к общей.
